```python
import random
```

```python
def ras(fn):
    def wrapper():
        frase = random.choice(['Ya es viernes', 'Hola'])
        return f"{frase} {fn()}"
    return wrapper
```

```python
@ras
def randomName():
    return random.choice(['Fer', 'Amanda'])
```
